```python
import pandas as pd
import numpy as np

# Load dataset
url = "https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv"
df = pd.read_csv(url)

# Display first 5 records
df.head()
```

```python
# Numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns
numerical_cols
```

```python
mean = df[numerical_cols].mean()
median = df[numerical_cols].median()
mode = df[numerical_cols].mode().iloc[0]

mean, median, mode
```

```python
minimum = df[numerical_cols].min()
maximum = df[numerical_cols].max()
variance = df[numerical_cols].var()
std_dev = df[numerical_cols].std()

minimum, maximum, variance, std_dev
```

```python
Q1 = df[numerical_cols].quantile(0.25)
Q2 = df[numerical_cols].quantile(0.50)  # Median
Q3 = df[numerical_cols].quantile(0.75)

Q1, Q2, Q3
```

```python
correlation = df[numerical_cols].corr()
correlation
```

```python
covariance = df[numerical_cols].cov()
covariance
```

```python
import matplotlib.pyplot as plt

df[numerical_cols].hist(figsize=(10,5))
plt.show()
```

```python
df[numerical_cols].boxplot(figsize=(8,5))
plt.show()
```
