```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Load Dataset (Heart Disease from UCI)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
df = pd.read_csv(url, names=columns, na_values='?')

# Data Cleaning (Required for plots to work)
df['sex'] = df['sex'].replace({0: 'Female', 1: 'Male'})
df['cp'] = df['cp'].replace({1: 'Typical Angina', 2: 'Atypical Angina', 3: 'Non-Anginal', 4: 'Asymptomatic'})
df.dropna(inplace=True)

# Set clearer style for printing
sns.set_style("whitegrid")

# --- Experiment 7: Line Plot & Bar Plot ---
plt.figure(figsize=(12, 5))

# Subplot 1: Line Plot (Trend)
plt.subplot(1, 2, 1)
plt.plot(df['trestbps'].head(20).values, marker='o', linestyle='-', color='b')
plt.title('Exp 7: Blood Pressure Trend (First 20 Patients)')
plt.xlabel('Patient ID')
plt.ylabel('Resting Blood Pressure (mm Hg)')

# Subplot 2: Bar Plot (Categorical Comparison)
plt.subplot(1, 2, 2)
sns.countplot(x='cp', data=df, palette='viridis')
plt.title('Exp 7: Count of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig('exp7_plots.png', dpi=300) # Save for printing
plt.show()

# --- Experiment 8: Scatter Plot ---
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='chol', hue='sex', data=df, palette='deep', s=60)
plt.title('Exp 8: Scatter Plot (Age vs Cholesterol)')
plt.xlabel('Age (Years)')
plt.ylabel('Cholesterol (mg/dl)')
plt.legend(title='Sex')
plt.tight_layout()
plt.savefig('exp8_scatter.png', dpi=300) # Save for printing
plt.show()

# --- Experiment 9: Histogram ---
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], kde=True, bins=20, color='purple')
plt.title('Exp 9: Age Distribution Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('exp9_histogram.png', dpi=300) # Save for printing
plt.show()

# --- Experiment 10: Boxplot ---
plt.figure(figsize=(8, 6))
sns.boxplot(x='chol', data=df, color='orange')
plt.title('Exp 10: Boxplot of Cholesterol Levels')
plt.xlabel('Cholesterol (mg/dl)')
plt.tight_layout()
plt.savefig('exp10_boxplot.png', dpi=300) # Save for printing
plt.show()
```
